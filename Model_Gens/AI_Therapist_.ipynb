```python
import shutil
import os

src = "/root/.cache/kagglehub/datasets/sujaykapadnis/emotion-recognition-dataset/versions/1"
dst = "/content/emotion-recognition-dataset"

# Use copytree if it's a folder
if os.path.exists(dst):
    print("Destination already exists, skipping copy.")
else:
    shutil.copytree(src, dst)
    print("Dataset copied successfully!")
```

```python
model=YOLO('yolov8n-cls.pt')
```

```python
import os, shutil, random

src = "/content/emotion-recognition-dataset/dataset"
train_dir = "/content/emotion-recognition-dataset/train"
val_dir = "/content/emotion-recognition-dataset/val"

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

for class_name in os.listdir(src):
    class_path = os.path.join(src, class_name)
    if not os.path.isdir(class_path):
        continue
    images = os.listdir(class_path)
    random.shuffle(images)
    split = int(0.8 * len(images))
    train_imgs = images[:split]
    val_imgs = images[split:]

    for img in train_imgs:
        os.makedirs(os.path.join(train_dir, class_name), exist_ok=True)
        shutil.copy(os.path.join(class_path, img), os.path.join(train_dir, class_name, img))
    for img in val_imgs:
        os.makedirs(os.path.join(val_dir, class_name), exist_ok=True)
        shutil.copy(os.path.join(class_path, img), os.path.join(val_dir, class_name, img))
```

```python
model.train(data="/content/emotion-recognition-dataset/train",epochs=50,batch=16)
```
